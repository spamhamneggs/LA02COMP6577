```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=["Alley", "MasVnrType", "MiscFeature","PoolQC","Fence",
                      "FireplaceQu"])
```

```python
num_features = df.select_dtypes(include=["int64", "float64"]).columns
cat_features = df.select_dtypes(include=["object"]).columns
num_features, cat_features
```

```python
from sklearn.impute import SimpleImputer

df[num_features] = SimpleImputer(strategy="median").fit_transform(df[num_features])
df[cat_features] = SimpleImputer(strategy="most_frequent").fit_transform(df[cat_features])
df.info()
```

```python
df = pd.get_dummies(df, columns=cat_features, dtype=int)
df.head()
```

```python
df = df.drop(columns=["Id"])
df.head()
```

```python
X = df.drop(columns=["SalePrice"])
Y = df["SalePrice"]
```

```python
from sklearn.model_selection import train_test_split

XTrain, XTest, YTrain, YTest = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(XTrain,YTrain)
```
